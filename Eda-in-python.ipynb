# **EDA in Python**


EDA is an approach to analysing data sets to summarize their main characteristics after using statistical graphics and other data visualisation methods.

It helps determine how best to manipulate data sources to get the answers we need , making it easier to spot anomalies , test a hypothesis or check assumption.



Seaborn is a data visualisation library in python. It helps to create interesting and interactive plots to draw meaningful information from data. Seaborn is build on top of matplotlib Data Visualisation library.

# **Bar Plot**

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
res = sns.barplot(mtcars["cyl"], mtcars["mpg"])
plt.show()
```

### ***Intution- As the cyl increases the mpg value decreases***

```python
res = sns.barplot(x="carb", y ="hp", data = mtcars, palette = "Paired" )
plt.show()
```

### *** Intution - Higher the numger of 'carb' more the hp***

```python
res = sns.barplot(x="am", y ="mpg", data = mtcars, palette = "Set2" )
plt.show()
```

### ***If am is 1 (manual) higher will be the mpg compared to 0(automatic)***

```python
res = sns.barplot(x="gear", y ="hp", data = mtcars, palette = "rocket" )
plt.show()
```

### ***Intution- Cars with gear = 4 have low horse power(hp)***

# **Count Plot**

A count plot is similar to a bar plot and a histogram and provides counts of categorical values

```python
sns.countplot(x = "cyl", data = mtcars, palette = "Set2")
plt.show()
```

```python
sns.countplot(x = "gear", data = mtcars)
plt.show()
```

```python
sns.countplot(x = "am", data = mtcars, palette ="Set1" )
plt.show()
```

# **Horizontal Count plots**

```python
sns.countplot(y = "carb", data = mtcars, palette ="Set1" )
plt.show()
```

# **Group count plot - Using hue parameter**

```python
# creating a group count plot that shows the no. of cars for each category of gears grouped by no.of cylinders
sns.countplot(x = "gear", hue = "cyl", data = mtcars )
```

### ***Majority of the car with 3 gears have 8 cyl and majority of car with 4 gears have 4 cyl***

```python
sns.countplot(x = "carb", hue = "cyl" , data = mtcars)
```

### ***Majority Cars with carb 1 and 2 have 4 cyl but as carb increases no. of cyl are either 6 or 8***

# **Dist plot**

A distplot plots a univariate distribution of observations. The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions.

```python
sns.distplot(mtcars.mpg, bins = 10, color = "g" )
```

### ***It is more likely to observe mpg near 20***

```python
sns.distplot(mtcars.qsec, bins = 10 ,color = "r")
```

### ***It is more likely to observe qsec near 18***

# **Heat Map**

A heatmap (or heat map) is a graphical representation of data where values are depicted by color.

```python
sns.heatmap(mtcars.corr(), cbar = True, linewidths = 0.5)
```

### ***Intution - Dark blue shows negative corelation and light colour shows positive corelaton***

# **Scatter Plot**

A scatter plot is a type of data visualization that shows the relationship between different variables. This data is shown by placing various data points between an x- and y-axis.

```python
sns.scatterplot(x = "disp", y = "hp", data = mtcars)
```

### ***Intution - hp and dist have an approximate linear relation***

```python
sns.scatterplot(x = "disp", y = "hp", data = mtcars , hue = "cyl" , size = "cyl")
```

### ***Using hue as cyl we can clearly see higher values of disp and hp have more number of cyl***

# **Joint Plot**

The joint plot is a way of understanding the relationship between two variables and the distribution of individuals of each variable

```python
sns.set()
sns.jointplot(x = "mpg", y = "disp" , data = mtcars, palette = "Set1", hue = "cyl" )
```

### ***Cars with 4 cyl have low value of disp and high value of mpg***

# **Pair plot**

Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters

```python
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(iris,hue = "species",palette = "Set1" )
```

### ***Intution - from the above pair plot we can find out that out of 4 features petal length is the best parameter to differentiate b/w diff species of flowers***

# **lmplot**

lmplots are basically scatter plots with overlaid regression lines

```python
sns.lmplot( x = "petal_length", y = "petal_width" ,data = iris)
```

```python
# other variation using hue parameter 
sns.lmplot( x = "petal_length", y = "petal_width", hue = "species",data = iris, markers= ["*","^","o"] )
```

# **Box And Whisker plot or Box plot**

A box and whisker plot is defined as a graphical method of displaying variation in a set of data. this plot detects the outliers if present.

```python
sns.boxplot( x = "species", y = "sepal_width", data = iris)
```

```python
sns.boxplot( x = "species", y = "petal_length", data = iris)
```

# **Swarm Plot**

A variation of box and whisker plot

```python
sns.swarmplot(x = "species", y = "sepal_width", data = iris)
```
